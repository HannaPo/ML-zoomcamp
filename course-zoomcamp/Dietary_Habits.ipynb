```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df =  pd.read_csv("project/Dietary Habits Survey Data.csv")
```

# Data preparation

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

string_columns = list(df.dtypes[df.dtypes == 'object'].index)

for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

```python
df.head().T
```

```python
df.drop('_', axis=1, inplace=True)
```

```python
df.columns = df.columns = ['age', 'gender', 'daily_count_meal', 'diet', 'skip_meals', 'feelings_of_hunger', 'consult_a_nutritionist', 'cook_my_own_meals', 'main_daily_meal', 'how_prepared', "orders", 'allergic', 'sweet_foods', 'salty_foods', 'fresh_fruit', 'fresh_vegetables','fried_foods', 'meat', 'seafood', 'tea', 'coffee','soft_drinks', 'juices', 'dairy_beverages', 'alcoholic_beverages', 'water' ]
```
